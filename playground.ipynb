```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    "occupation": [
        "casual-labour",
        "Employee",
        "Farmer",
        "Motorbike-rider",
        "Small-bussness",
        "Student"
    ],
    "percentage": [
        27.42,
        8.87,
        6.45,
        31.45,
        20.97,
        4.84
    ]
}
df = pd.DataFrame(data)

# Visualize with seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x="occupation", y="percentage", data=df, palette="viridis")
plt.title("Occupation Percentage Distribution")
plt.xlabel("Percentage")
plt.ylabel("Occupation")
plt.show()
```
